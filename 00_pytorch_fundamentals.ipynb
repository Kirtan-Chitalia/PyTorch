### Random Tensors

  why random tensors ?
- Random tensors are important because the way many neural network learn is that
  they start with full of random numbers and then adjust those random numbers
  to better represent the data

`Start with random numbers -> look at data -> update random numbers -> look at data -> update random numbers...`

## reshaping , stacking , squeezing and unsqueezing tensors

```python
# Creating a tensor
import torch
x = torch.arange(1.,10.)
x , x.shape
```

Reshaping - reshapes an input tensor to a defined shape
           ,Add an extra dimension


```python
x_reshaped = x.reshape(1,9)
x_reshaped , x_reshaped.shape
```

View - Returns a view of the original tensor in a different shape but shares the same data as the original tensor.

```python
z = x.view(1,9)
z , z.shape
```

```python
z[:,0] = 5
z,x
```

Stack - Concatenates a sequence of tensors along a new dimension (dim), all tensors must be same size.

```python
x_stacked = torch.stack([x,x,x,x], dim =0)
x_stacked
```

Squeeze - Squeezes input to remove all the dimenions with value 1

```python
print(f"Previous tensor: {x_reshaped}")
print(f"Previous shape: {x_reshaped.shape}")

# Remove extra dimension from x_reshaped
x_squeezed = x_reshaped.squeeze()
print(f"\nNew tensor: {x_squeezed}")
print(f"New shape: {x_squeezed.shape}")
```

Unsqueeze - Returns input with a dimension value of 1 added at dim.

```python
print(f"Previous tensor: {x_squeezed}")
print(f"Previous shape: {x_squeezed.shape}")

## Add an extra dimension with unsqueeze
x_unsqueezed = x_squeezed.unsqueeze(dim=0)
print(f"\nNew tensor: {x_unsqueezed}")
print(f"New shape: {x_unsqueezed.shape}")
```

Permute - Returns a view of the original input with its dimensions permuted (rearranged) to dims

```python
# Create tensor with specific shape
x_original = torch.rand(size=(224, 224, 3))

# Permute the original tensor to rearrange the axis order
x_permuted = x_original.permute(2, 0, 1) # shifts axis 0->1, 1->2, 2->0

print(f"Previous shape: {x_original.shape}")
print(f"New shape: {x_permuted.shape}")
```

```python
x_original[0,0,0] = 728218
x_original[0,0,0] , x_permuted[0,0,0]
```

# Indexing (Selecting data from tensors)

Indexing with Pythorch is Similar to Indexing in Numpy

```python
import torch
x = torch.arange(1,10).reshape(1,3,3)
x , x.shape
```

```python
#let's index on our new tensor
x[0]
```

```python
# Let's index on our middle bracket (dim=1)
x[0][0]
```

```python
# Let's index on our most inner bracket (dim=2)
x[0][0][0]
```

```python
# how to get number 9
x[0][2][2]
```

```python
# Get all values of 0th & 1st dimensions but only index 1 of 2nd dimension
x[:, :, 1]
```

```python
# Get all values of the 0 dimension but only the 1 index value of the 1st and 2nd dimension
x[:, 1, 1]
```

```python
# Get index 0 of 0th and 1st dimension and all values of 2nd dimension
x[0, 0, :] # same as x[0][0]
```

```python
# how to get 3,6,9
x[:, :, 2]
```
